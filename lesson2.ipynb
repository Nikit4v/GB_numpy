### Задание 2

```python
t = np.linspace(0, 10, 51)
```

```python
f = np.cos(t)
```

```python
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.plot(t, f)
plt.title("График f(t)")
plt.xlabel('Значения f')
plt.ylabel('Значения t')
```
